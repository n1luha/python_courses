```python
import numpy as np
import matplotlib.pyplot as plt
```

График 1

```python
x = np.linspace(2, 20, 5)

plt.figure(figsize=(10, 6))
plt.plot(x, np.cos(x), label='line 1', color='red', linewidth=2)
plt.plot(x, np.sin(x), '-.', label='line 2', color='green', linewidth=2)
plt.scatter(x, np.cos(x), color='red', s=70)
plt.scatter(x, np.sin(x), color='green', s=70)
plt.legend(loc='upper left')
plt.ylim(-1.5, 1.5)
```

Второй график

```python
grid = plt.GridSpec(2, 2)
x = np.linspace(1, 5, 5)
plt.figure(figsize=(10, 6))
plt.subplot(grid[0, :])
plt.plot(x, np.sin(x))
plt.subplot(grid[1, 0])
plt.plot(x, np.cos(x))
plt.subplot(grid[1, 1])
plt.plot(x, np.sin(x - np.pi / 2))
```

```python
x = np.linspace(-5, 5, 15)

plt.figure(figsize=(10, 6))
plt.plot(x, np.power(x, 2), linewidth=2)
plt.annotate('min', xy=(0, 0), xytext=(0, 10), arrowprops=dict(facecolor='green'))
```

График 4

```python
rng = np.random.default_rng(1)
data = rng.normal(5, 3, (7, 7))

plt.pcolormesh(data)
plt.colorbar(shrink = 0.5, aspect=7)
plt.clim(0, 10)
```

График 5

```python
x = np.linspace(0, 5, 1000)
plt.plot(x, np.cos(x * 3), color='red', linewidth=2)
plt.fill_between(x, np.cos(x * 3), color='darkblue', alpha=0.6)
```

График 6

```python
x = np.linspace(0, 5, 1000)
y = np.cos(x * 3)
yy = np.where(y >= -0.5, y, np.nan)
plt.plot(x, yy, linewidth=2)
plt.ylim(-1, 1)
```

График 7

```python
x = np.arange(0, 6, 1)
y = np.arange(0, 6, 1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.step(x, y, 'og-', where='pre')
plt.grid()

plt.subplot(1, 3, 2)
plt.step(x, y, 'og-', where='post')
plt.grid()

plt.subplot(1, 3, 3)
plt.step(x, y, 'og-', where='mid')
plt.grid()
```

График 8

```python
x = np.linspace(0, 10, 20)
y1 = -0.2 * (x - 5) ** 2 + 4
y2 = -0.6 * (x - 5) ** 2 + 12
y3 = -0.5 * (x - 7) ** 2 + 21
plt.fill_between(x, y1, label='y1')
plt.fill_between(x, y1, y2, label='y2')
plt.fill_between(x, y2, y3, label='y3')
plt.ylim(0, 25)
plt.legend(loc='upper left')
```
